```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

# 1. Prepare a list of offices sorted by country, state, city.

```python
run("SELECT officeCode, country, state, city FROM offices ORDER BY country, state, city")
```

# 2. How many employees are there in the company?

```python
run("SELECT count(CONCAT(firstName,' ',lastName)) AS Employees FROM employees")
```

```python
#atau
run("SELECT count(employeeNumber) AS Employees FROM employees")
```

# 3.	What is the total of payments received?

```python
run ("SELECT FORMAT(SUM(amount),2) AS Total_Payment FROM payments")
```

# 4. List the product lines that contain 'Cars'

```python
run ("SELECT productLine FROM productlines WHERE productLine LIKE '%cars%'")
```

# 5. Report total payments for October 28, 2004

```python
run("SELECT FORMAT(SUM(amount),2) AS Total_Payment FROM payments WHERE paymentDate='2004-10-28'")
```

# 6. Report those payments greater than $100,000.

```python
run("SELECT FORMAT(amount,2) AS Payment FROM payments WHERE amount>100000")
```

# 7.	List the products in each product line.

```python
run("SELECT productLine, productName FROM products ORDER BY productLine")
```

# 8. How many products in each product line?

```python
run("SELECT productLine, COUNT(productCode) AS Total FROM products GROUP BY productLine")
```

# 9.	What is the minimum payment received?

```python
run("SELECT MIN(amount) AS Minimum_Payment FROM payments")
```

# 10.	List all payments greater than twice the average payment.

```python
run("SELECT * FROM payments HAVING amount>(SELECT AVG(amount)*2 AS Rata2Harga FROM payments)")
```

# 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT productName, (MSRP-buyPrice) AS MarkUp FROM products ORDER BY productName")
```

```python
run("SELECT AVG(MSRP-buyPrice) AS average_markup FROM products")
```

# 14.	What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

# 15.	Which orders have a value greater than $5,000?
