# 1. Deskripsi Data

Dataset ini berisi tentang luas lahan, produksi, dan produktivitas dari tanaman kacang tanah per provinsi pada tahun 1993 sampai 2013. Data ini diambil dari data Badan Pusat Statistik (BPS).

Penjelasan mengenai Variabel pada Dataset ini:

- tahun: Tahun
- kode_provinsi: Kode Provinsi
- nama_provinsi: Nama Provinsi
- luas_panen: Luas Panen dari Tanaman Pangan (dalam Hektar)
- produktivitas: Produktivitas Tanaman Pangan (dalam Kuintal/Hektar)
- produksi: Total Produksi Tanaman Pangan (dalam Ton)
- latitude: koordinat lintang lokasi provinsi
- longitude: koordinat bujur lokasi provinsi

Luas bersih adalah luas sawah secara keseluruhan (luas kotor) dikurangi dengan luas pematang/galengan dan luas saluran air.

Luas panen adalah luasan tanaman yang dipungut hasilnya setelah tanaman tersebut cukup umur.

# 2. Statistik Sederhana tentang Data

```python
data_kacang.count()
```

```python
data_kacang.columns
```

```python
meluas = data_kacang.luas_panen.mean()
mepro = data_kacang.produktivitas.mean()
meproduksi = data_kacang.produksi.mean()

print('Rata-rata luas lahan panen tanaman kacang tanah adalah', meluas, 'hektar')
print('Rata-rata produktivitas tanaman kacang tanah adalah', mepro, 'kuintal/hektar')
print('Rata-rata produksi tanaman kacang tanah adalah', meproduksi, 'ton')
```

```python
sluas = data_kacang.luas_panen.sum()
spro = data_kacang.produktivitas.sum()
sproduksi = data_kacang.produksi.sum()

print('Jumlah luas lahan panen tanaman kacang tanah seluruhnya adalah', sluas, 'hektar')
print('Jumlah produktivitas tanaman kacang tanah seluruhnya adalah', spro, 'kuintal/hektar')
print('Jumlah produksi tanaman kacang tanah seluruhnya adalah', sproduksi, 'ton')
```

```python
miluas = data_kacang.luas_panen.min()
mipro = data_kacang.produktivitas.min()
miproduksi = data_kacang.produksi.min()

print('Nilai minimal dari luas lahan panen tanaman kacang tanah adalah', miluas, 'hektar')
print('Nilai minimal dari produktivitas tanaman kacang tanah adalah', mipro, 'kuintal/hektar')
print('Nilai minimal dari produksi tanaman kacang tanah adalah', miproduksi, 'ton')
```

# 3. Visualisasi Sederhana

# 3.1 Luas Lahan Panen per Provinsi

```python
ax = data_kacang.groupby('nama_provinsi').mean()['luas_panen']
ax = ax.plot(kind='bar', title ="Rata-rata Luas Lahan Panen", figsize=[10,5], legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Luas (hektar)", fontsize=12)
```

```python
data_kacang.columns
```

# 3.2 Jumlah Produktivitas Lahan Panen

```python
ax = data_kacang.groupby('nama_provinsi').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Rata-rata Jumlah Produktivitas", figsize=[10,5], legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah (kuintal/hektar)", fontsize=12)
```

# 3.3 Jumlah Produksi Kacang Tanah

```python
ax = data_kacang.groupby('nama_provinsi').mean()['produksi']
ax = ax.plot(kind='bar', title ="Rata-rata Jumlah Produksi Kacang Tanah", figsize=[10,5], legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah Produksi (ton)", fontsize=12)
```
